# preprocessing

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder(sparse_output=False) 
```

```python
encoded_col = encoder.fit_transform(df[['cb_person_default_on_file','person_home_ownership']])
```

```python
encoded_df = pd.DataFrame(encoded_col,columns=encoder.get_feature_names_out(['cb_person_default_on_file','person_home_ownership']))
```

```python
df = pd.concat([df,encoded_df],axis=1)
df.drop(['cb_person_default_on_file','person_home_ownership','id'],axis=1,inplace=True)
```

```python
df.head()
```

```python
label_encoder = {}
for i,j in enumerate(df.select_dtypes(include='object').columns):
    a,b = pd.factorize(df[j])
    label_encoder[j] = {y:x for x,y in zip(set(a),b)}
```

```python
df.replace(label_encoder, inplace=True)
```

```python
X = df.drop('loan_status',axis=1)
y = df['loan_status']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=.2,random_state=1)
```

# model development

```python
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
model = XGBClassifier(random_state=1)
```

```python
model.fit(X_train,y_train)
```

```python
pred = model.predict(X_test)
```

```python
print(classification_report(y_test,pred))
```
